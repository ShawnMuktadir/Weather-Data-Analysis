Q. 19) What is the average humidity for each weather condition?

```python
data.groupby('Weather')['Rel Hum_%'].mean().sort_values().plot(kind='barh')
```

```python
# Set a nicer style
sns.set(style='whitegrid')

# Create the plot
plt.figure(figsize=(12, 8))
colors = sns.color_palette('coolwarm', len(data['Weather'].unique()))

data.groupby('Weather')['Rel Hum_%']\
    .mean()\
    .sort_values()\
    .plot(kind='barh', color=colors)

# Add titles and labels
plt.title('Average Relative Humidity by Weather Condition', fontsize=16, fontweight='bold')
plt.xlabel('Average Humidity (%)', fontsize=12)
plt.ylabel('Weather Condition', fontsize=12)

# Add grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Make layout tight
plt.tight_layout()

# Show the plot
plt.show()
```

Q. 20) What is the relationship between temperature and humidity?

Q. 21) How does wind speed vary with weather conditions?

```python
#Chart: Box Plot
#Purpose: Compare distribution across categories.

# Set the figure size (e.g., 18 inches wide by 6 inches tall)
plt.figure(figsize=(18, 8))
sns.boxplot(x='Weather', y='Wind Speed_km/h', data=data)
# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Set labels and title (optional)
plt.xlabel('Weather')
plt.ylabel('Wind Speed (km/h)')
plt.title('Wind Speed Variation by Weather Condition')

# Show the plot
plt.tight_layout()
plt.show()
```

Q. 22) What is the trend of average daily temperature over months?

```python
#Chart: Line Chart (after resampling)
#Purpose: Long-term trend.
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
#Month-End frequency ('ME' stands for “Month End”), meaning each group ends at the last day of each month.
monthly_temp = data.set_index('Date/Time').resample('ME')['Temp_C'].mean()
monthly_temp.plot(title='Monthly Avg Temp')
```
